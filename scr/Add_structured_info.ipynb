```python
# get all inclang2doc files and merge
docs_to_find = ['2015 Challenger Banque Nationale de Drummondville',
                'ATP-toernooi van Indianapolis 1999',
                '1999 RCA Championships',
                'Indianapolis Tennis Championships 1999,']

inc2doc_merged = defaultdict(dict)

paths_inc2doc = glob.glob('../data/releases-and-repos-sorted/*/structured/inc2lang2doc_index.json')
print(len(paths_inc2doc))
for path in paths_inc2doc:
    print(path)
    with open(path) as infile:
        inc2doc = json.load(infile)
    for inc, lang_docs in inc2doc.items():
        # print(inc)
        for lang, docs in lang_docs.items():
            # Process docs
            # print(len(docs))
            for doc in docs:
                if '/' in doc:
                    doc = doc.split('/')[-1]
                doc = doc.encode('utf-8')
                doc = doc.decode('utf-8')
                if lang in inc2doc_merged[inc]:
                    if doc not in inc2doc_merged[inc][lang]:
                        inc2doc_merged[inc][lang].append(doc)
                    else:
                        continue

                else:
                    inc2doc_merged[inc][lang] = [doc]
                    
# paths_inc2doc2 = glob.glob('../data/releases-and-repos-sorted/*/structured/inc2doc_index.json')
# for path in paths_inc2doc2:
path = '../data/releases-and-repos-sorted/v1/structured/inc2doc_index.json'
print(path)
with open(path) as infile:
    inc2doc_v1 = json.load(infile)
# print(inc2doc_v1)
for inc, lang_docs_v1 in inc2doc_v1.items():
    # print(inc)
    if inc == 'Q19594339':
        print('found inc')
        # print(lang_docs_v1)
        # print()
    for lang_doc in lang_docs_v1:
        lang, docs = lang_doc.split('/')

        for doc in docs:
            doc = doc.encode('utf-8')
            doc = doc.decode('utf-8')
            if inc == 'Q19594339':
                print('found inc again', inc, path)
                print(doc)
            if doc in docs_to_find:
                print('found', doc)
            if lang in inc2doc_merged[inc]:
                if doc not in inc2doc_merged[inc][lang]:
                    inc2doc_merged[inc][lang].append(doc)
                else:
                    continue

            else:
                inc2doc_merged[inc][lang] = [doc]
```

```python
# get all type2inc files and merge

type2inc_merged = defaultdict(list)

paths_type2inc = glob.glob('../data/releases-and-repos-sorted/*/structured/type2inc_index.json')


for path in paths_type2inc:
    print(path)
    with open(path) as infile:
        type2inc = json.load(infile) 
    for type_q, incs in type2inc.items():

        for inc in incs:
            if inc not in type2inc_merged[type_q]:
                type2inc_merged[type_q].append(inc)
            else:
                continue
    
    
```

```python
inc2type = dict()
for inc_type, incs in type2inc_merged.items():
    for inc in incs:
        inc2type[inc] = inc_type
```

```python
#inc2doc
title_structured_dict = defaultdict(dict)

for inc, lang_docs in inc2doc_merged.items():
    #print(inc)
    if inc in inc2type:
        type_q = inc2type[inc]
    else:
        type_q = '-'
    for lang, docs in lang_docs.items():
        # print(len(docs))
        # docs = set(docs)
        # print(len(docs))
        # print(docs)
        for title in docs:
            title = title.encode('utf-8')
            title = title.decode('utf-8')
            title_structured_dict[title]['inc_q'] = inc
            title_structured_dict[title]['type_q'] = type_q
```

```python

all_docs = set()

for inc, lang_docs in inc2doc_merged.items():
    for lang, docs in lang_docs.items():
        all_docs.update(docs)
                
# check

with open('text-titles-from-structured.txt', 'w') as outfile:
    for doc in sorted(all_docs):
        outfile.write(f'{doc}\n')
    
```

```python
#inc2doc

for d in data:
    lang = d['lang']
    title = d['text_title'].split('.')[0].encode('utf-8').decode('utf-8')
    structured_d = title_structured_dict[title]
    if len(structured_d) > 0:
        type_q = structured_d['type_q']
        inc_q = structured_d['inc_q']
        d['type_q'] = type_q
        d['inc_q'] = inc_q 
    # else:
    #     print('not found')
    #     print(title)
```
