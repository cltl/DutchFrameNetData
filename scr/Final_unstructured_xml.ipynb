# Finalize xml dataset

```python
import csv
from lxml import etree as et
from lxml.etree import XMLSyntaxError
```

```python
with open('../data/overview-unique-ids-structured.csv') as infile:
    data = list(csv.DictReader(infile, delimiter=','))
```

```python
naf_files = '../data/data-unique-ids/unstructured'
naf_files_new = '../data/data-unique-ids-final/unstructured'
```

```python
for d in data:
    lang = d['lang']
    uuid = d['unique_id']
    path = f'{naf_files}/{lang}/{uuid}.naf'
    
    path_new = f'{naf_files_new}/{uuid}.naf'
    
    with open(path) as infile:
        naf = infile.read()
        
    # check if valid naf
    
    # try:
    try:
        tree = et.parse(path)
        with open(path_new, 'w') as outfile:
            outfile.write(naf)
        d['valid_naf'] = 'TRUE'
    except XMLSyntaxError:
        d['valid_naf'] = 'FALSE'
        with open(path_new, 'w') as outfile:
            outfile.write(naf)
        
        

          
```

```python
header = data[0].keys()
with open('../data/overview-unique-ids-structured.csv', 'w') as outfile:
    writer = csv.DictWriter(outfile, fieldnames = header, delimiter = ',')
    writer.writeheader()
    for d in data:
        writer.writerow(d)
```
