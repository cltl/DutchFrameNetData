# Convert XML to JSON

```python
import xmltodict
import json
import os
```

```python
json_dir = "../data/data-unique-ids-final/unstructuredJSON"
#os.mkdir(json_dir)
```

```python
folder_path = "../data/data-unique-ids-final/unstructured"
files = [f for f in os.listdir(folder_path) if f.endswith(".naf")]

for file in files[:10]:
    xlm_filepath = os.path.join(folder_path, file)
    json_filepath = os.path.join(json_dir, file+".json")
    print(file)
    print(json_filepath)
    with open(xlm_filepath) as xml_file:
        data_dict = xmltodict.parse(xml_file.read())
    #json_data = json.dumps(data_dict)
    with open(json_filepath, "w") as json_file:
        json.dump(data_dict, json_file)
```
