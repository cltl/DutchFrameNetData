```python
from lxml import etree as et
import csv
import os
```

```python
# check data

with open('../data/overview-unique-ids.csv') as infile:
    data = list(csv.DictReader(infile))
```

```python
# find manually annotated file
for d in data:
    uu = d['unique_id']
    lang = d['lang']
    ann = d['annotation_mode']
    path = f'../data/data-unique-ids/unstructured/{lang}/{uu}.naf'
    if os.path.isfile(path):
        if ann == 'system-manual':
            print('manual annotations')
            print(path)
           
    
    else:
        print('problem')
        print(path)
        print(d)
        print()
```

```python
# change path

path = '../data/data-unique-ids/unstructured/en/3016b308-7b28-4968-b25e-399a6b97e554.naf'
tree = et.parse(path)
root = tree.getroot()
```

```python
# expore

for child in root.getchildren():
    print(child.tag)
```

```python
# go directly to srl

srl_layer = root.find('srl')
srl_layer.tag
```

```python
# explore srl layer

for child in srl_layer.getchildren():
    print(child.tag)
```

```python
# get all predicates (in a list)
predicates = srl_layer.findall('predicate')
```

```python
# look at a single predicate
print(et.tostring(predicate, pretty_print=True).decode())
```

```python
for predicate in predicates:
    status = predicate.get('status')
    if status== 'manual':
        print(predicate.tag, predicate.attrib)
        break
```

```python
# get information from one predicate

frame_info = predicate.findall('externalReferences/externalRef')
for ref in frame_info:
    print(ref.tag, ref.attrib)

# get span
span = predicate.findall('span/target')
for sp in span:
    print(sp.tag, sp.attrib)
    
    
# get roles:
roles = predicate.findall('role')
for role in roles:
    print(role.tag, role.attrib)
```

```python
# inspect role element
print(et.tostring(roles[0], pretty_print=True).decode())
```
